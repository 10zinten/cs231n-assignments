```python
from scipy.misc import imread, imsave, imresize
```

## Read an JPEG image into a numpy array

```python
img = imread('assets/cat.jpeg')
print(img.dtype, img.shape)
```

## Tint the image

```python
img_tinted = img * [1, 0.95, 0.9]  # Leaving red channel unchanged
```

## Resize the Tinted Image

```python
img_tinted = imresize(img_tinted, (300, 300))
```

## Save the Tinted Image to disk 

```python
imsave('assets/cat_tinted.jpeg', img_tinted)
```

## Showing image with matplotlib

```python
import matplotlib.pyplot as plt 
import numpy as np
```

```python
# Show the original image
plt.subplot(1, 2, 1)
plt.imshow(img)

# Show the tinted image
plt.subplot(1, 2, 2)
plt.imshow(np.uint8(img_tinted))
plt.show()
```
